```python
# пункт 1, 2, 5 и 6
import yfinance as yf
import pandas as pd

crypto = ['BTC-USD', 'ETH-USD', 'BCH-USD', 'BNB-USD', 'LINK-USD', 'LTC-USD', 'DOT1-USD', 'XTZ-USD', 'ADA-USD', 'ATOM1-USD']
df = pd.DataFrame(columns=crypto)
for currency in crypto:
    df[currency] = yf.download(currency, period = '6mo')['Adj Close']
df
```

```python
# пункт 3
# предположим, что стоимость криптовалюты зависит от динамики курса доллара к евро

import matplotlib.pyplot as plt

USD = yf.download('USDEUR=X', period = '6mo')['Adj Close']
%matplotlib inline
USD.plot()
plt.show()
```
